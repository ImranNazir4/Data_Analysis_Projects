```python
import pandas as pd
import numpy as np
import re
import neattext as nt
import os
```

```python
files=os.listdir()
```

```python
df=pd.read_csv("zara.csv")
```

```python
def extract_user_handles(tweet):
    pattern = r'@(\w+)'  # Regular expression pattern to match Twitter handles
    handles = re.findall(pattern, tweet)
    return ', '.join(handles) if handles else None
```

```python
def extract_hashtags(tweet):
    pattern = r'#(\w+)'  # Regular expression pattern to match hashtags
    hashtags = re.findall(pattern, tweet)
    return ', '.join(hashtags) if hashtags else None
```

```python
def clean_tweet_text(tweet):
    # Remove URLs
    tweet = re.sub(r'http\S+|www\S+|https\S+', '', tweet, flags=re.MULTILINE)
    
    # Remove user mentions (@username)
    tweet = re.sub(r'@\w+', '', tweet)
    
    # Remove hashtags (#hashtag)
    tweet = re.sub(r'#\w+', '', tweet)
    
    # Remove special characters and punctuations
    tweet = re.sub(r'[^\w\s]', '', tweet)
    
    # Remove extra spaces
    tweet = re.sub(r'\s+', ' ', tweet).strip()
    
    return tweet
```

```python
files=[i for i in files if ".ipynb" not in i]
```

```python
files
```

```python
dfs=[]
```

```python
for i in files[1:]:
    df=pd.read_csv(i)
    df["Username"]=df["Username"].apply(lambda x: x.split("\n")[0])
    df["Date"]=df["Username"].apply(lambda x: x.split("\n")[-1])
    df["UserHandles"]=df["Tweet"].apply(lambda x:extract_user_handles(str(x)))
    df["Hastags"]=df["Tweet"].apply(lambda x:extract_hashtags(str(x)))
    df["Tweet_Len"]=df["Tweet"].str.len()
    df["Cleaned_Tweet"]=df["Tweet"].apply(lambda x:clean_tweet_text(str(x)))
    dfs.append(df)
```

```python
pd.concat(dfs)
```
