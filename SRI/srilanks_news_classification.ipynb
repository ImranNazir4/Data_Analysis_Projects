```python
import pandas as pd
import numpy as np
import neattext as nt
from nltk.tokenize import word_tokenize #importing word tokenizer from NLTK
from nltk.stem import WordNetLemmatizer    #importing wordnet lemmatizer from NLTK
lemmatizer = WordNetLemmatizer()  #creating object of lemmatizer
```

```python
import nltk
nltk.download('wordnet')     #downloading dependency for text lemmatization
nltk.download('averaged_perceptron_tagger')  # Download the necessary dataset for POS tagging
nltk.download('punkt')   #downloading tokenizers
```

```python
def lemmatization(text):
    """
    this functin recieves text from text preprocessing method and return lemmatized text
    """
    words = text.split()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    lemmatized_sentence = ' '.join(lemmatized_words)
    return lemmatized_sentence
```

```python
def text_preprocessing(text):
    text=nt.fix_contractions(text)     #I'm -> I am
    text=nt.remove_stopwords(text)     #removing stopwords
    text=nt.remove_emojis(text)     #removing stopwords
    text=nt.remove_urls(text)          #removing urls
    text=nt.remove_non_ascii(text)     #removing non-ascii characters
    text=nt.remove_userhandles(text)   #removing urserhandles
    text=nt.remove_hashtags(text)      #removing hashtags
    text=nt.remove_multiple_spaces(text)  #removing multiple spaces
    text=lemmatization(text)           #lemmatization
    return text
```

```python
df["cleaned_text"]=df["text"].apply(lambda x:text_preprocessing(x))
```
