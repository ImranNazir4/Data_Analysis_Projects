### Column Description

This dataset contains 22 columns and the column description is given below:
1. **trans_date_trans_time**: Transaction date and time 
2. **cc_num**: Credit card number 
3. **merchant**: Merchant at which fraud occured 
4. **category**: Category of merchant 
5. **amt**: Amount of credit card fraud 
6. **first**: First name of the card holder 
7. **last**: Last name of the card holder 
8. **gender**: Gender of the card holder 
9. **street**: Street of the crad holder 
10. **city**: City of the card holder 
11. **state**: State of the card holder 
12. **zip**: Zip code of the card holder 
13. **lat**: Latitude 
14. **long**: Longitude 
15. **city_pop**: Population of the city of card holder 
16. **job**: Job type of the card holder 
17. **dob**: Date of birth of card holder 
18. **trans_num**: Transaction number 
19. **unix_time**: The unix time stamp is a way to track time as a running total of seconds. This count starts at the Unix Epoch on January 1st, 1970 at UTC.
20. **merch_lat**: Merchant's latitude 
21. **merch_long**: Merchant's longitude 
22. **is_fraud**: Is it a fraud transaction or not?
