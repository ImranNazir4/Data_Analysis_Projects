```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
v1=pd.read_excel('anscombe_data_v1.xlsx')
v2=pd.read_excel('anscombe_data_v2.xlsx')
v3=pd.read_excel('anscombe_data_v3.xlsx')
v4=pd.read_excel('anscombe_data_v4.xlsx')
```

# I

```python
sns.lmplot(x="x",y="y",data=v1,ci=0)
print("X mean",np.mean(v1.x))
print("Y mean",np.mean(v1.y))
print("X std",np.std(v1.x))
print("Y std",np.std(v1.y))
```
