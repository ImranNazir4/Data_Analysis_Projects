```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('FaucetBathroom_webdata.xls')
```

```python
columns=df.columns
```

```python
for i in columns:
    print(i)
```

```python
single_value_cols=[]
multi_value_cols=[]
```

```python
for i in multi_value_cols:
    print(i,' ----------- ',df[i].isnull().sum()/df.shape[0])
```

```python
df[multi_value_cols[0:4]]
```

```python
df[multi_value_cols]
```

```python
for i in columns:
    if 'Multi' not in i:
        single_value_cols.append(i)
    else:
        multi_value_cols.append(i)
        print(i)
```

```python
single_value_df=df[single_value_cols]
```

```python
null_percent_over_60=[]
null_percent_over_50=[]
null_percent_over_30=[]
null_percent_over_20=[]
```

```python
for i in single_value_cols:
    null_percent=df[i].isnull().sum()/df.shape[0]
    if null_percent>=0.20:
        null_percent_over_20.append(i)
    if null_percent>=0.30:
        null_percent_over_30.append(i)
    if null_percent>=0.50:
        null_percent_over_50.append(i)
    if null_percent>=0.60:
        null_percent_over_60.append(i)
    print(i,' --------------------- ',null_percent)
```

```python
null_percent_over_60
```

```python
null_percent_over_50
```

```python
single_value_df=single_value_df.drop(null_percent_over_50,axis=1)
```

```python
for i in single_value_df.columns:
    print(i,' ----------------- ',single_value_df[i].isnull().sum(),'/',single_value_df.shape[0])
```

```python
single_value_df
```

```python
single_value_df.info()
```

```python
single_value_df_imputed=single_value_df.copy()
```

```python
single_value_df_imputed
```

```python
for i in single_value_df.columns:
    single_value_df_imputed.fillna(single_value_df[i].mode()[0],inplace=True)
```

```python

for i in single_value_df.columns:
   # plt.figure(figsize=(10,10))
    fig, ax = plt.subplots(figsize=(25, 10))
    
    sns.distplot(single_value_df[i], hist=True, kde=True, ax=ax,color='red')
    sns.distplot(single_value_df_imputed[i], hist=True, kde=True, ax=ax, color='green')
    plt.title(i)
    plt.show()
```

```python
df=pd.concat([single_value_df_imputed,df[multi_value_cols]],axis=1)
```

```python
df.to_csv('FaucetBathroom_webdata.csv')
```
