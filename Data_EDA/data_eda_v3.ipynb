### Storages Analysis

```python
#this function converts peta bytes to tera bytes and also remove trailing data units
def pb_to_tb(petabytes):
    if petabytes[-1]=='P':
        petabytes=petabytes[0:-1]
        petabytes=float(petabytes)
        terabytes=petabytes * 1024
        return terabytes
    else: 
        return petabytes[1:-1]
```

```python
df['Used HDD']=df['Used HDD'].apply(lambda x:str(x).split()[0])
```

```python
df['Used HDD']=df['Used HDD'].apply(lambda x:pb_to_tb(str(x)))
```

```python
df['Used  SSD Storage']=df['Used  SSD Storage'].apply(lambda x:str(x).split()[0])
```

```python
df['Used  SSD Storage']=df['Used  SSD Storage'].apply(lambda x:pb_to_tb(str(x)))
```

```python
df['Avail HDD']=df['Avail HDD'].apply(lambda x:str(x).split()[0])
df['Avail HDD']=df['Avail HDD'].apply(lambda x:pb_to_tb(str(x)))
```

```python
df['Avail  SSD Storage']=df['Avail  SSD Storage'].apply(lambda x:str(x).split()[0])
df['Avail  SSD Storage']=df['Avail  SSD Storage'].apply(lambda x:pb_to_tb(str(x)))
```

```python
df['VHS Size HDD']=df['VHS Size HDD'].apply(lambda x:pb_to_tb(str(x)))
```

```python
#cnverting string data to numerical data
for i in ['Used HDD','Used  SSD Storage', 'Avail HDD', 'Avail  SSD Storage','VHS Size HDD']:
    df[df[i]=='']=np.nan
    df[df[i]=='/a']=np.nan
    df[i]=df[i].astype(float)
```

```python
for i in ['Used HDD','Used  SSD Storage', 'Avail HDD', 'Avail  SSD Storage','VHS Size HDD']:
    sns.displot(df[i])
```

```python
df[df['Hottest_HDD_Temp']=='na']=0
df[df['Hottest_HDD_Temp']=='no reading']=0
df['Hottest_HDD_Temp']=df['Hottest_HDD_Temp'].astype(float)
sns.displot(df['Hottest_HDD_Temp'])
plt.title('Hottest HDD Temperature')
```

### Ram Analysis

```python
df['RAM']=df['RAM'].apply(lambda x:str(x).split()[0])
```

```python
df['RAM']=df['RAM'].astype(float)
```

```python
#converting bytes to gb
def bytes_to_gb(byte):
    gb = byte / (1024 ** 3)
    return gb
```
