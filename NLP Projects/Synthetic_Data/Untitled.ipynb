```python
import pandas as pd
```

```python
df=pd.read_excel('OPENAIGOOD.xlsx')
```

```python
df.shape
```

```python
df['label']='None'
```

```python
df['label'][0:100]='None'
```

```python
df.dropna(inplace=True)
```

```python
df['label'][100:]='Not None'
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE


# Define the TF-IDF vectorizer
tfidf = TfidfVectorizer()

# Convert the text data to feature vectors
X = tfidf.fit_transform(df['Open_AI_Text'])
y = df['label']

# Define the SMOTE resampler
smote = SMOTE(sampling_strategy=5.0, random_state=42)

# Apply SMOTE to the feature vectors
X_resampled, y_resampled = smote.fit_resample(X, y)

# Convert the synthetic feature vectors back to text
synthetic_text = tfidf.inverse_transform(X_resampled)

# Write the synthetic text to a new CSV file
pd.DataFrame(synthetic_text, columns=['Open_AI_Text']).to_csv('synthetic_data.csv', index=False)
Ke
```
