```python
import pandas as pd
import numpy as  np
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
df_list=[]
```

```python
def entity_entractor(text):
    docx=nlp(text)
    ents_list=[(ent.text,ent.label_) for ent in docx.ents]
    df=pd.DataFrame(ents_list,columns=["Entity","Tag"])
    df_list.append(df)
```

```python
data=pd.read_excel('property_comments.xlsx')
```

```python
data['comment'].apply(lambda x:entity_entractor(x))
```

```python
pd.concat(df_list,axis=0)
```
